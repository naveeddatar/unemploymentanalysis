#### d)
Finally, compare the two cryptocurrencies. You first need to merge the two dataframes based on the date. <br>

```python
btc_eth = pd.merge(btc.add_suffix("_BTC"), eth.add_suffix("_ETH"), left_index=True, right_index=True)
btc_eth
```

Create a column showing the difference in closing prices. What is the averge, min, max difference?

```python
btc_eth["Close_diff"] = btc_eth["Close_BTC"] - btc_eth["Close_ETH"]
print("Average distance:", btc_eth["Close_diff"].mean())
print("Min distance:", btc_eth["Close_diff"].min())
print("Max distance:", btc_eth["Close_diff"].max())
```

Create a DataFrame showing the relative increase in closing prices. Answer the following questions:
- In how many % of days did the prices BTC and ETH increase or decrease?
- In how many % of days was the change in BTC price exceed the one of ETH?

*Hint: You can get the number of rows via `len(df.index)`.*

```python
btc_eth_incr = (btc_eth - btc_eth.shift()) / btc_eth.shift()
btc_eth_incr = btc_eth_incr[["Close_BTC", "Close_ETH"]].dropna()
btc_eth_incr
```

```python
btc_up = len(btc_eth_incr[btc_eth_incr["Close_BTC"]>0].index)/len(btc_eth_incr.index)*100
btc_down = len(btc_eth_incr[btc_eth_incr["Close_BTC"]<0].index)/len(btc_eth_incr.index)*100
eth_up = len(btc_eth_incr[btc_eth_incr["Close_ETH"]>0].index)/len(btc_eth_incr.index)*100
eth_down = len(btc_eth_incr[btc_eth_incr["Close_ETH"]<0].index)/len(btc_eth_incr.index)*100
print("BTC price increased in {}% of days.".format(round(btc_up,2)))
print("BTC price decreased in {}% of days.".format(round(btc_down,2)))
print("ETH price increased in {}% of days.".format(round(eth_up,2)))
print("ETH price decreased in {}% of days.".format(round(eth_down,2)))
```

```python
btc_eth_incr["BTC_above_ETH"] = (btc_eth_incr["Close_BTC"] > btc_eth_incr["Close_ETH"])
btc_above_eth = sum(btc_eth_incr["BTC_above_ETH"])/len(btc_eth_incr.index)*100
print("BTC price change above ETH price change in {}% of days.".format(round(btc_above_eth,2)))
```

## Problem 2

In this problem, you will take a look at FitBit fitness tracker data (taken from Kaggle: https://www.kaggle.com/datasets/gloriarc/fitbit-fitness-tracker-data-capstone-project).

```python
fitbit = pd.read_csv("Daily_Activity_2022_27_02.csv")
fitbit.head()
```

```python
# Get rid of last two columns.
fitbit = fitbit.drop(columns=["Unnamed: 16", "Unnamed: 17"])
```

#### a)
Set a multi-index with ID, date, and day.

```python
# Turn ActivityDate into datetype and set index
fitbit["ActivityDate"] = pd.to_datetime(fitbit["ActivityDate"])
fitbit = fitbit.set_index(["Id", "ActivityDate", "ActivityDay"])
fitbit
```

#### b)
Find the average daily distance and total calories per individual.

```python
fitbit_grouped = fitbit.groupby(["Id"])
fitbit_grouped.agg({"TotalDistance": "mean",
                    "Calories": "sum"})
```

#### c)
Construct the same DataFrame using `pivot_table`. 

```python
fitbit_wide = pd.read_csv("Daily_Activity_2022_27_02.csv")
fitbit_wide["ActivityDate"] = pd.to_datetime(fitbit_wide["ActivityDate"])
columns_of_interest = ["TotalSteps", "TotalDistance", "VeryActiveDistance", "ModeratelyActiveDistance", "LightActiveDistance", "SedentaryActiveDistance", "VeryActiveMinutes", "FairlyActiveMinutes", "LightlyActiveMinutes", "SedentaryMinutes", "Calories"]
fitbit_wide = fitbit_wide.pivot_table(
    index="ActivityDate",
    columns=["Id"],
    values=columns_of_interest
)
fitbit_wide
```

```python
means = fitbit_wide.mean()
sums = fitbit_wide.mean()
pd.concat([means.loc["TotalDistance", :], sums.loc["Calories", :]], axis=1).rename(columns={0: "TotalDistance", 1: "Calories"})
```
